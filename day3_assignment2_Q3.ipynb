# Question 3
 Write a program for printing the tables from 1,10 using Nested For Loop

```python
for num1 in range(1,11):                          # putting value in num1 1 to 10 
    print("")                                     # just a space between tables
    print("Table of",num1)                        # printing the number of which table is to be printed
    for num2 in range(1,11):                      # looping from 1 to 10 to get table 
        print (num1,"*",num2,"=",num1*num2);      # printing num1 * num2 = num1*num2
```
